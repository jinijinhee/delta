```python
##client 클라이언트

import socket
import time
 

# Create a client socket

clientSocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM);

 

# Connect to the server

clientSocket.connect(("127.0.0.1",6027));

 

# Send data to server
a = [];
data = "Hello Server!";

clientSocket.send(data.encode());

 

# Receive data from server
while True:
    time.sleep(1)
    dataFromServer = clientSocket.recv(1024).decode('utf-8', 'ignore');
    if dataFromServer == 'bye':
        print(dataFromServer)
        break
    a.append(dataFromServer)
    print(dataFromServer)
```
